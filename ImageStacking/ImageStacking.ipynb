## Horizondal and Vertical Stacking

```python
import cv2
import numpy as np

folder_path = "/content/drive/My Drive/ML_Seizure_Project/InputImageForStacking/034images"
if os.path.exists(folder_path):
    file_list = os.listdir(folder_path)


    file_list = [os.path.join(folder_path, x) for x in file_list if x.endswith(('.jpg', '.png', '.bmp', '.jpeg'))]

else:
    print("The specified folder does not exist.")
```

```python
count=0
for file in file_list:
  print(file)
  image = cv2.imread(file)

  image=cv2.resize(image,(300,300))

  if count==0:
    HorizontalStack1=np.hstack([image])
    VerticalStack1=np.vstack([image])
    count=count+1

  else:
    HorizontalStack1=np.hstack([HorizontalStack1,image])
    VerticalStack1=np.vstack([VerticalStack1,image])
```

## Horizontal stacking

```python
cv2_imshow(HorizontalStack1)
```
